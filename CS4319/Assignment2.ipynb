```python
import numpy as np
import tabulate as tabulate

testData = np.loadtxt('testdata.txt', delimiter=',')
testClass = np.loadtxt('test_class.txt', dtype = int)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    x = np.exp(x - np.max(x))
    return x / x.sum()

def forwardPass(data, weights):
    J = np.dot(data, np.array([[weights['w1'], weights['w2']], [weights['w3'], weights['w4']]]))

    H = relu(J)

    I = np.dot(H, np.array([[weights['w5'], weights['w6']], [weights['w7'], weights['w8']]]))

    prediction = np.apply_along_axis(softmax, 1, I)

    return prediction

def accuracy(predictions, labels):
    predictedLabels = np.argmax(predictions, axis=1)
    return np.mean(predictedLabels == labels)

weights1 = {'w1': 1.3438, 'w2': -0.6225, 'w3': 0.3509, 'w4': -1.7072, 'w5': -1.1398, 'w6': 0.3944, 'w7': 1.3882, 'w8': -0.8676}
weights2 = {'w1': 0.8061, 'w2': 0.2354, 'w3': -0.4092, 'w4': -0.8999, 'w5': -0.5538, 'w6': -0.1916, 'w7': 0.0288, 'w8': 0.4918}

predictions1 = forwardPass(testData, weights1)
predictions2 = forwardPass(testData, weights2)

headers = ["Sample", "Prediction Model 1", "Prediction Model 2", "Actual Class"]
table_data = [
    [i+1, np.argmax(predictions1[i]), np.argmax(predictions2[i]), testClass[i]]
    for i in range(len(testClass))
]

print(tabulate.tabulate(table_data, headers=headers, tablefmt="grid"))

accuracy1 = accuracy(predictions1, testClass)
accuracy2 = accuracy(predictions2, testClass)

print(f'Accuracy of model 1: {accuracy1} or {accuracy1*100:.4f} %')
print(f'Accuracy of model 2: {accuracy2} or {accuracy2*100:.4f} %')

if accuracy1 > accuracy2:
    print('Model 1 is better')
elif accuracy2 > accuracy1:
    print('Model 2 is better')
else:
    print('Both models are equally good')
```
