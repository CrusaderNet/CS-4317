# Let's compute statistics for each text in Gutenberg Corpus using for loop

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

for i in gutenberg.fileids():
    num_chars = len(gutenberg.raw(i))          # unprocessed text
    number_words = len(gutenberg.words(i))     # Tokenized into words
    num_sents = len(gutenberg.sents(i))        # Tokenized into sentences
    num_vocab = len(set([w.lower() for w in gutenberg.words(i)]))  # unique words
    print(int(num_chars/number_words),"  ",int(number_words/num_sents),"  ",int(number_words/num_vocab),"  ", i)
```

# The first column: average word length
# The second column: average sentence length (mean no. of words per sentence)
# The third column: the number of times each vocabulary item apperas in the text on average (our lexical diversity)

**Note: The average word length is really 3, not 4, since the num_chars variable counts space characters too.**

# Conditional Frequency Distribution
**conditional frequency distribution is a way to organize and analyze the frequency of occurrences of words in a given text or corpus, with respect to some condition or context.**

```python
cfd = nltk.ConditionalFreqDist(
                                (genre, word)
                                for genre in brown.categories()
                                for word in brown.words(categories=genre))

genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']

cfd.tabulate(conditions=genres, samples=modals)
```

**Observe that the most frequent modal in the news genre is will, while the most frequent modal in the romance genre is could.**

**udhr, contains the Universal Declaration of Human Rights in over 300 languages.**

**Let’s use a conditional frequency distribution to examine the differences in word lengths for a selection of languages included in the udhr corpus.**

```python
from nltk.corpus import udhr
import matplotlib.pyplot as plt
languages = ['Chickasaw', 'English', 'German_Deutsch',
            'Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
cfd = nltk.ConditionalFreqDist(
           (lang, len(word))
           for lang in languages
           for word in udhr.words(lang + '-Latin1'))
plt.figure(figsize=(10,8))
cfd.plot(cumulative=False)
```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)
```

**Our program iterates through every word and, for each one, checks whether it meets the conditions. It is easy to check obligatory letter and length constraints (and we’ll only look for words with six or more letters here). It is trickier to check that candidate solutions only**
**use combinations of the supplied letters, especially since some of the supplied letters
appear twice (here, the letter v). The FreqDist comparison method permits us to
check that the frequency of each letter in the candidate word is less than or equal to the
frequency of the corresponding letter in the puzzle.**

```python
puzzle_letters = nltk.FreqDist('egivrvonl')
```

```python
puzzle_letters
```

```python
import nltk
puzzle_letters = nltk.FreqDist('egivrvonl')
print("Frequency Distribution:",end="\n\n")
for i in puzzle_letters:
    print(i,"  ",puzzle_letters[i])
    
obligatory = 'r'
wordlist = nltk.corpus.words.words()
words_after_computation = [w for w in wordlist if len(w) >= 7
                    and obligatory in w
                    and nltk.FreqDist(w) <= puzzle_letters]

words_after_computation = set(words_after_computation)
```
