# One Hot Encoding

```python
########################################
#
#What is the output of the following program? What kind of matrix is it? What kind of Encoding is it?
#
########################################
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

### Categorical data to be converted to numeric data
colors = (["red", "green", "yellow", "red", "blue"])

### integer mapping using LabelEncoder
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(colors)
print(integer_encoded)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)

### One hot encoding
onehot_encoder = OneHotEncoder(sparse_output=False)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)

print(onehot_encoded)
```

# Cosine Similarity

# Angle Between Vectors	Cosine Similarity	Interpretation
# 0° ->1 -> Identical vectors (same direction)
# 90°->0->Completely unrelated (orthogonal vectors)
# 180°->-1->Opposite vectors (exact opposite direction)
# 360°->1->Same as 0°, vectors are identical

# The result is the cosine of the angle between the two vectors, ranging from  -1 to 1. A value of 1 indicates perfect similarity, 0 indicates orthogonality, and -1 indicates perfect dissimilarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
A = np.array([7,3])
B = np.array([3,7])

cos_sim = cosine_similarity(A.reshape(1,-1),B.
                            reshape(1,-1))
print(f"cosine similarity is {cos_sim}")
```

### Normalized Term Frequency

- Above, using `bag of words` or frequency distribution we can describe a document. 
- A document can also be described as a `vector` containing the normalized frequencies of the unique words.
- `normalized term frequency` is `TF` divided by total number of tokens in a text.
- Instead of using TF, We use normalized TF of a word to measure the importance of a word in a text. 

`normalized TF`

```python
for col in df.columns:
    df[col] = df[col]/df.sum(axis=1)
df
```

Important words in `df`

```python
df.sort_values(by=['doc_trump'],axis=1,ascending=False)
```

- `IDF(t,D) = log(no. of docs/no. of docs containing t) + 1`

IDF=log(1)=0
which would eliminate the term’s contribution entirely. Adding +1 ensures that even frequently occurring words retain some weight.


The scikit-learn `TFIDF` matrix (not normalized is) for the first `doc` is

```python
import math
tmp1 = np.array([1*(math.log((4+1)/(4+1))+1), 0, 1*(math.log((4+1)/(2+1))+1),  1*(math.log((4+1)/(2+1))+1), 0])
```

```python
from numpy.linalg import norm

norm(tmp1)
```
